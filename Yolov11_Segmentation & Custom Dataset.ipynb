```python
!pip install ultralytics
```

```python
#Trainig Custom Dataset
from ultralytics import YOLO

# Load a YOLOv11 segmentation model (or use 'yolo11l-seg.pt' as base)
model = YOLO('yolo11l-seg.pt')  # Or 'yolo11s-seg.pt' etc.

# Train on custom data
model.train(
    data='/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/data/pothole.v2i.yolov11/data.yaml',  # Path to your data.yaml
    epochs=100,
    imgsz=640,
    device='cuda',      # or 'cpu'
    project='/content/drive/MyDrive/Yolov11_Segmentation/Yolov11_Custom_PotHole/results',
    name='custom_seg_train',
    batch=2,            # Adjust based on your system memory
    workers=2,
    patience=20,
    save=True,
    save_period=1
)
```

```python
#Validation
from ultralytics import YOLO

# Load your trained model
model = YOLO('/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/results/custom_seg_train/weights/best.pt')

# Run validation
model.val(
    data='/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/data/pothole.v2i.yolov11/data.yaml',  # same as training
    imgsz=640,
    device='cuda',  # or 'cpu'
    split='val',    # validates on the validation set
    save=True,      # saves results
    project='/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/val_results',  # result dir
    name='best_model_eval'  # subfolder name
)
```

```python
from ultralytics import YOLO

best_model = YOLO('/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/results/custom_seg_train/weights/best.pt')

#Segmentation Detection
best_model.predict(
    source="/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/inference/pothole.mp4",
    imgsz=640,
    conf=0.25,
    device='cpu',
    project="/content/drive/MyDrive/Yolov11_Segmentation2/Yolov11_Custom_PotHole/results",
    name="Seg-result",
    save=True,
    show=True,
    line_width=1  # For masks, thinner lines are often better
)
```
